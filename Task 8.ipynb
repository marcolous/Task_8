```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('smoke_detection_iot.csv')
```

```python
df
```

```python
count = df['Fire_Alarm'].value_counts()
```

```python
plt.pie(count, autopct='%1.1f%%', wedgeprops={'linewidth': 1, 'edgecolor': 'black'}, startangle= 180)
colors = plt.cm.get_cmap('tab20').colors[:len(count)]
labels = ['Yes_Fire', 'No_Fire']
plt.legend(labels, loc='center', bbox_to_anchor=(1.5,0.7))
plt.show()
```
